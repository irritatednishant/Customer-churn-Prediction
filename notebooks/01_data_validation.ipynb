# 01 Data Validation
Validate the raw Telco customer churn dataset before analysis or modeling.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
DATA_PATH = "../data/raw/Telco_customer_churn.xlsx"

df = pd.read_excel(DATA_PATH)
df.head()
```

```python
print("Shape:", df.shape)
df.dtypes
```

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values[missing_values > 0]
```

```python
print("Duplicate rows:", df.duplicated().sum())
print("Duplicate CustomerID:", df['CustomerID'].duplicated().sum())
```

```python
df.sample(7, random_state=42)
```

```python
print("Non-null Total Charges:", df['Total Charges'].notnull().sum())

total_charges_numeric = pd.to_numeric(df['Total Charges'], errors='coerce')
print("Total Charges values that become missing after numeric conversion:", total_charges_numeric.isnull().sum())

df.loc[total_charges_numeric.isnull(), ['CustomerID', 'Tenure Months', 'Monthly Charges', 'Total Charges']].head()
```

```python
print("Missing Churn Reason values:", df['Churn Reason'].isnull().sum())
print("Missing Churn Reason by Churn Label:")
print(df.groupby('Churn Label')['Churn Reason'].apply(lambda s: s.isnull().sum()))
```

```python
for col in ['Tech Support', 'Paperless Billing', 'Contract', 'Internet Service', 'Payment Method']:
    print(f"\n===== {col} =====")
    print(df[col].unique())
```

```python
validation_summary = pd.DataFrame({
    'dtype': df.dtypes.astype(str),
    'missing_count': df.isnull().sum(),
    'missing_percent': df.isnull().mean() * 100,
    'unique_values': df.nunique()
})
validation_summary
```
